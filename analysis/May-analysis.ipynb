```python
import pandas as pd
fn = "results.xlsx"
f = pd.read_excel(fn, index_col=0, sheet_name="Tab1-data")
```

```python
opts = {"S": 12.55, "C": 10.61, "A": 6445}
```

```python
def format_gap(gap):
    if pd.isna(gap):
        return "?"
    left, right, pert = "{", "}", "\%"
    strs = f"{gap*100:.1f}".split(".")
    return f"{strs[0]}{left}\scriptsize.{strs[1]}{pert}{right}"
```

```python
to_concat = []
for task in opts:
    x = f[f["task"] == task].iloc[:,1:]
    x = (x - opts[task]) / opts[task]
    y = x.applymap(format_gap)
    to_concat.append(y)

g = pd.concat(to_concat, axis=1)
g.to_excel("gaps_formatted.xlsx")
```
